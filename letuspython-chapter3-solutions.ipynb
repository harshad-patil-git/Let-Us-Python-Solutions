#### Assume suitable value of temperature of a city in Fahrenheit degrees. Write a program to convert this temp. into Centigrade degrees and print both temperatures.

```python
f_temp = 70
c_temp = (f_temp - 32) * (5/9)
print('Temperature in Fahrenheit: ', f_temp)
print('Temperature in Centigrade: ', c_temp)
```

#### Given three sides a,b,c of a triangle, write a program to obtain and print the values of three angles rounded to the next integer.

```python
a = 5
b = 6
c = 4
```

```python
A = math.ceil(math.acos((b**2 + c**2 - a**2)/(2*b*c)))
B = math.ceil(math.acos((a**2 + c**2 - b**2)/(2*a*c)))
C = math.ceil(math.acos((a**2 + b**2 - c**2)/(2*a*b)))
```

```python
print('Angle A: ', A)
print('Angle B: ', B)
print('Angle C: ', C)
```
